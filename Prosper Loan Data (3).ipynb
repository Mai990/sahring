```python
# import needed libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb

%matplotlib inline
```

```python
# loading data
loan_data = pd.read_csv('prosperLoanData.csv')
df=loan_data
```

```python
#See the first rows
loan_data.head()
```

```python
print(df.columns)
```

Starting with univariate

```python
# What's the Distribution of Borrower Rate
plt.figure(figsize=[14.70, 8.27])
plt.hist(data= df, x= "BorrowerRate", edgecolor = "white");
plt.title('Distribution of Borrower Rate', y=1.05, fontsize=16, fontweight='bold', color = 'Blue');
plt.xlabel("Borrower Rate")
plt.ylabel('Count')
```

The number of loans issued to borrowers is higher for 0.25 (which is 25% in this dataset) than for 0.1 (or 10 %). The number loans issued for rates higher than 25% falls after that

```python
#States with most borrowers

plt.figure(figsize=[14.70, 10.27])
sb.countplot(data = df, y = "BorrowerState", color = "blue");
```

California is the highest state with big different between it and other states. We can see Florida,New York and Texas are in the second place.

```python
#Lets find out what's the Estimated Loss

plt.figure(figsize=[14.70, 8.27]);
plt.hist(data= df, x= "EstimatedLoss" , edgecolor = "white");
plt.title('Estimated Loss', y=1.05, fontsize=16, fontweight='bold', color = 'Blue');
plt.xlabel('Estimated Loss')
plt.ylabel('Count')
```

The estimated loss percentile is getting higher from 0.01 (or 10%) till 0.15 which is 15% in this dataset

```python
# What's the Distribution of Percent Funded
plt.figure(figsize=[14.70, 8.27])
plt.hist(data= df, x= "BorrowerRate", edgecolor = "white");
plt.title('Distribution of Percent Funded', y=1.05, fontsize=16, fontweight='bold', color = 'Blue');
plt.xlabel("Percent Funded")
plt.ylabel('Count')
```

The percentage of the funded starts from 0.1 (or 10%) and gets higher till reaching 0.2 ( or 20%) then decreasing till 0.39 (or 39%)

Bivariante 

```python
#Who are more luckly to have loan, let's describe their employment status
plt.figure(figsize=[14.70, 8.27]);
sb.countplot(data = df, y = "EmploymentStatus", color = "blue");

```

A huge number of employed people are taking loans, with big difference comes the people who work full-time. on the opposite, the retired and the people who work in part-time jobs are way more less luckly to take loan. 

```python
#Let's find out the relation between the EmploymentStatus and the LoanOriginalAmount
plt.figure(figsize=[14.70, 8.27])
sb.boxplot(data= df, x="LoanOriginalAmount", y="EmploymentStatus");
```

it showes that the employed person are most luckly to take higher loans. Then the self-employed in the second place.

```python
#Now what about the Employment satus duration
plt.figure(figsize=[14.70, 9.27]);
sb.countplot(data = df, x = "EmploymentStatusDuration", color = "blue");
```

```python
#does the Employment satus duration affects the status of the loan
plt.figure(figsize=[14.70, 8.27])
sb.boxplot(data= df, x="EmploymentStatusDuration", y="LoanStatus");
```

When the employment status duration increases the more completing the loan increases too

```python
#We have seen the borrower rate above but Does the Income range affects it
plt.figure(figsize=[14.70, 8.27])
sb.boxplot(data= df, x="BorrowerRate", y="IncomeRange");
plt.ylabel('IncomeRange (in USD)')
```

Non employed people are the most people who borrow, then the employed ones with income ranges from $1 to aprox $25,000
